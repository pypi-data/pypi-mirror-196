```python
%%sql 
select code,
date,
AVG(C) OVER (PARTITION BY code ORDER BY date ROWS BETWEEN 4 PRECEDING AND CURRENT ROW) AS day_avg
FROM stock.parquet
```

```python
%%sql 
SELECT
code,
date,
100 - (100 / (1 + RS)) AS rsi
FROM (
SELECT
code,
date,
AVG(U) / AVG(D) AS RS
FROM (
SELECT
code,
date,
CASE WHEN C > LAG(C, 1) OVER (PARTITION BY code ORDER BY date) THEN C - LAG(C, 1) OVER (PARTITION BY code ORDER BY date) ELSE 0 END AS U,
CASE WHEN C < LAG(C, 1) OVER (PARTITION BY code ORDER BY date) THEN LAG(C, 1) OVER (PARTITION BY code ORDER BY date) - C ELSE 0 END AS D
FROM stock.parquet
) t
GROUP BY code, date
HAVING COUNT(*) >= 14
) t2;
```
